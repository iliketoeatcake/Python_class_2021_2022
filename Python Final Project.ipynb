```python
import csv
filename="pokemon.csv"
d={}  
# opening the file using "with"
# statement
with open(filename,'r') as csv_file:
    csv_reader=csv.reader(csv_file,delimiter=',')
    line_count=0
    for row in csv_reader:
        if line_count==0:
            print(f'Column names are {".".join(row)}')
            line_count+=1
        else:
            d[row[1]]=row[12]        # create a dictionary , use name (row[0] as a key and University  row[7] as a value)      
            if line_count<6:        # Line 15/16 are helping you to print the table before line 5
                print(f'{".".join(row)}')
            line_count+=1
           
            
    print(f'Processed {line_count} lines.')
    print(d)
```

```python
def reverse_lookup_list(d, v): # two arguements need:, one is the dictonary and the other is the value
    res=[]
    for i in d.keys(): # i in d may also work.
        if d[i]==v:
            res.append(i)
    return res

reverse_lookup_list(d, 'True') #To find all legendaries
```
